# R benchmarks

```r
library(tictoc)
```

```r
set.seed(42)
```

## linear

```r
n = 1000000; k = 20
X = matrix(rnorm(n*k), n, k)
(b = runif(k))
y = X %*% b + rnorm(n)

tic()
(m = lm(y ~ -1 + X)$coefficients)
toc()
```

## Logistic

```r
n = 10000; k = 20
X = matrix(rnorm(n*k), n, k)
(b = runif(k))
y = rbinom(n,1, plogis(X %*% b))

tic()
(m = glm(y ~ -1 + X, family=binomial())$coefficients)
toc()
```

## Poisson

```r
n = 100000; k = 10
X = matrix(rnorm(n*k), n, k)
(b = runif(k))
y = rpois(n, exp(X %*% b))

tic()
(m = glm(y ~ -1 + X, family=poisson())$coefficients)
toc()
```
